Instance Method
*   First parameter = self (always)
*   Can access class + instance attributes *

Class Method
*   First parameter = class cls
*   Can access only classs attributes **
*   Decorator = @classmethod (when we want to convert any method into a class method

Static Method
*   No compulsory parameter
*   Can not access instance attributes, nor class attributes
*   Static Method Decorator - normal functions ko static method me convert krta h
*   Classes ke sath linked hote hn


```python
class Laptop:
  type_of_laptop="SSD"

  def __init__(self,ram,storage):
    self.ram=ram
    self.storage=storage

  @classmethod #is function ka behaviour change krdia - is function ko class ka method bana dia.
  def get_storage_type(cls):
    print(f"This laptop has {cls.type_of_laptop}.")

  def get_info(self):
    print(F"This laptop has {self.ram} of RAM and {self.storage} of storage, and is of {self.type_of_laptop} type.") #*

  @staticmethod
  def calc_discount(price,discount):
    final_price=price-(discount*price/100)
    print(f"Final price = {final_price}.")

l1=Laptop("8GB","128GB")
l2=Laptop("16GB","256GB")

print(f"{l1.ram}, {l1.storage}")
l1.get_info()
l1.get_storage_type() #**
Laptop.get_storage_type()

l1.calc_discount(40_000, 10)
#fnx => (price, discount) => final price
#this function is neither using any of the instance parameters, nor the class parameters #but bcz iska logic is related to classes
```

```python
'''question
Design & create an online product store which stores the products(name,price).
Track total products being created.
Create a static method to calculate discount on each product based on % parameter.
'''

#step1: Decide INSTANCE attribute or CLASS attribute (common for all or unique for all)
#name & price - unique - instance attributes
#count variable - jo sari objects ka count store kare - jaise hi object bane, count variable update ho jaye - common for all - class attribute

class Product:
  count=0 #class attribute

  def __init__(self,name,price):
    self.name=name
    self.price=price
    Product.count +=1

  def get_info(self):
    print(f"The price of {self.name} is {self.price}.")

  @classmethod
  def get_count(cls):
    print(f"Total products created = {cls.count}.")

  def calc_discount(self,discount):
    final_price=self.price-(discount*self.price/100)
    print(f"Final price of {self.name} = {final_price}.")


p1=Product("Phone",20_000)
p2=Product("Laptop",50_000)
p3=Product("Earphone",10_000)

p1.get_info()
p2.get_info()
p3.get_info()

Product.get_count()

p1.calc_discount(10)
p2.calc_discount(20)
p3.calc_discount(15)
```

INHERITANCE

Har class ki properties hoti hn. Properties being inherited from parent class to child class (derived class).

```python
class School:
  start="8am"
  end="5pm"
  def change_time(self,new_end):
    self.end=new_end

class Teacher(School): #teacher ke jitne bhi objects honge unke pas by default school ke properties chali jayengin:
  def __init__(self,subject):
    self.subject=subject

class AdminStaff(School):
  def __init__(self,name,role):
    self.name=name
    self.role=role

  def get_info_admin(self):
    print(f"Role of {self.name} is {self.role}.")

a1=AdminStaff("Rahul","HR")
a2=AdminStaff("Meena","Manager")

t1=Teacher("Maths")
t2=Teacher("Science")

t1.change_time("6pm")

print(f"{t1.subject},{t1.start},{t1.end}")
print(f"{t2.subject},{t2.start},{t2.end}")

a1.get_info_admin()
a2.get_info_admin()

#parent class me koi aisa attribute h jise ham SUB-CLASS ko bhi dena chahte hn + PRIVATE bhi rkhna chahte hn (class ke bahar access nhi, but sub-class ko dena chahte hn - PROTECTED)
##parent class me koi aisa attribute h jise ham SUB-CLASS ko bhi NHI dena chahte hn + PRIVATE bhi rkhna chahte hn (class ke bahar access nhi, sub-class ko bhi NHI - PRIVATE)
```

1. Single Level Inheritance - one parent class. one child class.
2. Multi Level Inheritance - one parent class. then sub-class, sub-sub class, sub-sub-sub class (hierarchy)
3. Multiple Inheritance - 2 or more parent classes

```python
class School:
  start="8am"
  end="5pm"
  def change_time(self,new_end):
    self.end=new_end

class Teacher(School): #teacher ke jitne bhi objects honge unke pas by default school ke properties chali jayengin:
  def __init__(self,subject):
    self.subject=subject

class AdminStaff(School):
  def __init__(self,name,role):
    self.name=name
    self.role=role

class Accountants(AdminStaff): #automatically SCHOOL ki + ADMIN STAFF dono ki properties aa jayengin
  def __init__(self,name,salary,role):
    super().__init__(name, role) #to call the contructor of a parent class with both name and role
    self.salary=salary

account1=Accountants("John", 50000,"CA")
print(account1.salary, account1.start, account1.name, account1.role)
```

```python
class Teacher:
  def __init__(self,salary):
    self.salary=salary

class Student:
  def __init__(self,gpa):
    self.gpa=gpa

class TA(Teacher,Student):
  def __init__(self,salary,gpa,name):
    super().__init__(salary)
    Student.__init__(self,gpa)
    self.name=name

ta1=TA(15000,9.0,"Rahul")
print(ta1.salary,ta1.gpa,ta1.name)
```

Abstract classes- inse ham dusri classes create krte hn, abc module abstraction based classes module me already likhi hui h python me, bas import krke use krna hota h. Inke khudke koi obkjects nhi hote. Iske methods ko ham child classes me implement krte hn.

```python
from abc import ABC, abstractmethod

class Animal(ABC):
  @abstractmethod
  def make_sound(): #abstract method
    pass #represents null value - abhi is function ke andar koi kaam perform nhi hoga

class Lion(Animal):
  def make_sound(self):
    print("Roar!")
class Cat(Animal):
  def make_sound(self):
    print("Meow!")

l1=Lion()
l1.make_sound()
c1=Cat()
c1.make_sound()
```
