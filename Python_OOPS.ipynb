Suppose a school has a blank admission form. Student Form

Name: ________
Age: ________

This blank form is your class.

"Create one student and fill in the name as Rahul." So what does __init__() do?

Think of __init__() as the person who fills the form.

def __init__(self, name):

means "Whenever a new Student is created, I need a name."

-------------------------------------------------------------------------------
self.name = name

means "Take the name that was given and write it inside this student's record."

You write stu1 = Student("Rahul")

Python automatically does this behind the scenes:

A new Student object is created.

↓

Python calls __init__()

↓

Inside __init__(): name = "Rahul"

↓

self.name = name    becomes

This student's name = Rahul

Done!

Now another student comes. cycle repeats

SELF simply means "this student"

When you write (self.name = name) you're saying "Store the name inside this student."

```python
#method

class Student:
    def __init__(self, name, cgpa):
        self.name = name
        self.cgpa=cgpa
    def hello(self):
        print("Hello", self.name)
    def welcome(self):
      print("Welcome to SRCC,",self.name,"!")
    def get_cgpa(self):                                                         #ham ek fn banana chahte hn jo hame kisi bhi student ki cgpa return krke de
      return self.cgpa

#create object
stu1 = Student("Rahul", 9.09)
stu2 = Student("Sohan", 8.5)
stu3 = Student("Rohit",9.8)

print(stu1.name)
print(stu2.name)
print(stu3.name,",",stu3.cgpa)

#call method
stu1.hello()
stu1.welcome()
stu1.get_cgpa()
print(f"CGPA of {stu1.name} is {stu1.cgpa}.")
```
