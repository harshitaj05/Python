```python
#string FORMATTING - we want to create dynamic strings (we use different variables and values in the string)
#2 methods: 1. FORMAT() function (placeholder, placement value)

a=5
b=10
sum=a+b

#normal formatting
print("sum of {} & {} is {}".format(a,b,sum))
print("language is {}".format("python"))

#index based formatting
print("sum of {1} & {0} is {2}".format(a,b,sum))

#value based formatting - to be used when we want to reuse variables
print("{a}values of variables {a} & {b}".format(a=5,b=10))
```

```python
'''
.append(value) #add one element at the end
.insert(index,value) #add element at the particular index
.sort() #increasing order
.reserve() #reverses the order
'''

marks=[99,100,80,91,78]

marks.append(76)
print(marks)

#INCREASING order
marks.sort()
print(marks)

#DECREASING order
marks.sort(reverse=True)
print(marks)

marks.reverse()
print(marks)
```

```python
#Input two lists from user, merge them into one, then sort them.

my_list1=[]
my_list2=[]

number1=input("Enter your numbers separated by space:").split()
number2=input("Enter your numbers separated by space:").split()

for i in number1:
  my_list1.append(i)

for j in number2:
  my_list2.append(j)

combined_list = my_list1+my_list2
print(combined_list)

combined_list.sort()
print(combined_list)
```
