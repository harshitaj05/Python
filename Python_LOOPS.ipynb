```python
print("hello world") #100x

#while - kaam ko tabtak karenge jab tak it is giving a true value
'''
while "a condition":
  print(")

infinite loop (till the memory fills up, tab tak print krta rahega)
while "a condition which is always true":
  print(")
ctrl+C to stop it
'''

count=1 #ITERATOR, i, j, k

while count<=10:  #terminating condition
  print("Loop is here x5.",count)
  count +=1

print("After loop, count value=",count)

#generally, we start the forward loop iterator always with 0 (count = 0). This is because some index number.
```

```python
#PRINT BACK COUNT FROM 10 TO 0

count2=10
while count2>=0:
  print(count2)
  count2 -=1
```

```python
#print the table of 6

number=1

while number<=10:
  print(6*number)
  number +=1
```
